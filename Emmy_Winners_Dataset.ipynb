```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import requests

url = "https://imdb188.p.rapidapi.com/api/v1/emmyWinners"

headers = {
	"X-RapidAPI-Key": "YOUR API KEY",
	"X-RapidAPI-Host": "imdb188.p.rapidapi.com"
}

response = requests.get(url, headers=headers)
```

```python
df = response.json()
x = df['data']
x
```

```python
x = x['list']
x
```

```python
x[0]['title']
```

```python
df = pd.DataFrame(x)
df
```

```python
for index,key in df['title'].items():
    keys_in_df = key.keys()
    print(keys_in_df)
```

```python
def return_list(data_f,Key_to_check):
    lst = []
    
    for index,key in data_f.items():
        keys_in_df = key.keys()
        #print(keys_in_df)
        
        if Key_to_check in keys_in_df:
            lst.append(key[Key_to_check])
    return lst

# Basic cols made
lst_receiver = return_list(df['title'],'id')
df['id'] = lst_receiver  


lst_receiver = return_list(df['title'],'titleText')
df['titleText'] = lst_receiver  


lst_receiver = return_list(df['title'],'releaseYear')
df['releaseYear'] = lst_receiver


lst_receiver = return_list(df['title'],'titleType')
df['titleType'] = lst_receiver


lst_receiver = return_list(df['title'],'ratingsSummary')
df['ratingsSummary'] = lst_receiver


lst_receiver = return_list(df['title'],'titleEpisode')
df['titleEpisode'] = lst_receiver


lst_receiver = return_list(df['title'],'isAdult')
df['isAdult'] = lst_receiver

# Further filtering the columns made 

lst_receiver = return_list(df['titleText'],'text')
df['titleText'] = lst_receiver  

lst_receiver = return_list(df['releaseYear'],'year')
df['Release_Year'] = lst_receiver

lst_receiver = return_list(df['releaseYear'],'endYear')
df['End_Year'] = lst_receiver
df['End_Year'].replace(np.nan, 0, inplace=True) # removing NaN
df['End_Year'] = df['End_Year'].astype('int') #changing float values of this column to int 


lst_receiver = return_list(df['titleType'],'text')
df['Title_Type'] = lst_receiver

lst_receiver = return_list(df['titleType'],'canHaveEpisodes')
df['can_Have_Episodes'] = lst_receiver

lst_receiver = return_list(df['titleType'],'isSeries')
df['is_Series'] = lst_receiver

# ratingsSummary

lst_receiver = return_list(df['ratingsSummary'],'aggregateRating')
df['Aggregate_Rating'] = lst_receiver

lst_receiver = return_list(df['ratingsSummary'],'voteCount')
df['Vote_Count'] = lst_receiver

lst_receiver = return_list(df['ratingsSummary'],'topRanking')
df['topRanking'] = lst_receiver

lst_receiver = return_list(df['topRanking'],'rank')
df['topRanking'] = lst_receiver

df
```

```python
df = df.drop(['title','releaseYear','titleType','ratingsSummary','titleEpisode','can_Have_Episodes'], axis=1)
df
```

```python
df = df.reindex(columns=['id', 'titleText', 'Title_Type','topRanking','Aggregate_Rating','Vote_Count','Release_Year','End_Year','isAdult','is_Series'])
df
```

```python
df.to_csv('Series_details.csv',index=False)
```
